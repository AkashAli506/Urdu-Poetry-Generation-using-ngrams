##Preprocessing on iqbal.txt, ghalib.txt and faiz.txt


```python
words=[]
for sentence in iqbal:
  if not sentence.isspace():
    sentence = sentence.strip()
    sentence = sentence.replace('.','')
    sentence = sentence.replace(',','')
    sentence = sentence.replace(':','')
    sentence = sentence.replace(';','')
    sentence = sentence.replace('!','')
    sentence = sentence.replace('?','')
    sentence = sentence.replace('(','')
    sentence = sentence.replace(')','')
    sentence = sentence.replace("-"," ")
    sentence = sentence.replace("' ",'')
    sentence = sentence.replace('"','')
    sentence = sentence.replace("_","")
    sentence = sentence.replace("“","")
    word = sentence.split(' ')
    words += word

for sentence1 in ghalib:
  if not sentence1.isspace():
    sentence1 = sentence1.strip()
    sentence1 = sentence1.replace('.','')
    sentence1 = sentence1.replace(',','')
    sentence1 = sentence1.replace(':','')
    sentence1 = sentence1.replace(';','')
    sentence1 = sentence1.replace('!','')
    sentence1 = sentence1.replace('?','')
    sentence1 = sentence1.replace('(','')
    sentence1 = sentence1.replace(')','')
    sentence1 = sentence1.replace("-"," ")
    sentence1 = sentence1.replace("' ",'')
    sentence1 = sentence1.replace('"','')
    sentence1 = sentence1.replace("_","")
    sentence1 = sentence1.replace("“","")
    word1 = sentence1.split(' ')
    words += word1

for sentence1 in faiz:
  if not sentence1.isspace():
    sentence1 = sentence1.strip()
    sentence1 = sentence1.replace('.','')
    sentence1 = sentence1.replace(',','')
    sentence1 = sentence1.replace(':','')
    sentence1 = sentence1.replace(';','')
    sentence1 = sentence1.replace('!','')
    sentence1 = sentence1.replace('?','')
    sentence1 = sentence1.replace('(','')
    sentence1 = sentence1.replace(')','')
    sentence1 = sentence1.replace("-"," ")
    sentence1 = sentence1.replace("' ",'')
    sentence1 = sentence1.replace('"','')
    sentence1 = sentence1.replace("_","")
    sentence1 = sentence1.replace("“","")
    word1 = sentence1.split(' ')
    words += word1


words = [word.strip('’’') for word in words]
words = [word.strip(' ') for word in words]
words = [word.strip('‘‘') for word in words]
words = [word.strip('') for word in words]

wordList=list()
count=0
for i in range(0,len(words)):
    if (words[i]==''):
      continue
    wordList.append(words[i])
print(len(wordList))
```

#Count Unigrams and their probabilities

```python
uniCounts = dict({})
uniGramProbs = dict({})
for i in range(0,len(wordList)):
  if wordList[i] in uniCounts:
      uniCounts[wordList[i]]+=1
  else:
      uniCounts[wordList[i]]=1
for word in uniCounts:
  prob=(uniCounts.get(word)/sum(uniCounts.values()))
  uniGramProbs[word]=prob
print(len(uniGramProbs))
```

#Count Bigrams and their probabilities

```python
biCounts= dict({})
biGramProbs= dict({})
for i in range(0,len(wordList)-1):
  word1=wordList[i]
  word2=wordList[i+1]
  bigram= word1+" "+word2
  if bigram in biCounts:
    biCounts[bigram]+=1
  else:
    biCounts[bigram]=1 
for myword in biCounts:
  splitWords=myword.split(" ")
  firstWord=splitWords[0]
  prob=(biCounts.get(myword)/uniCounts.get(firstWord))
  biGramProbs[myword]=prob
print(len(biGramProbs))
```

#Count Trigrams and their probabilities

```python
triCounts= dict({})
triGramProbs= dict({})
for i in range(0,len(wordList)-2):
  word1=wordList[i]
  word2=wordList[i+1]
  word3=wordList[i+2]
  trigram= word1+" "+word2+ " "+word3
  if trigram in triCounts:
    triCounts[trigram]+=1
  else:
    triCounts[trigram]=1 
for triWord in triCounts:
  myBiGram=triWord.split(" ")
  myNewBiGram=myBiGram[0]+" "+myBiGram[1]
  probNewBiGram=biCounts.get(myNewBiGram)
  if (probNewBiGram==None):
    continue
  else:
    totalProb=(triCounts.get(triWord)/probNewBiGram)
    triGramProbs[triWord]=totalProb
print(len(triGramProbs))
```

```python
backwardBiCounts= dict({})
backwardBiGramProbs= dict({})
i=len(wordList)
while(i>1):
  word1=wordList[i-1]
  word2=wordList[i-2]
  backwardBigram= word1+" "+word2
  if backwardBigram in backwardBiCounts:
    backwardBiCounts[backwardBigram]+=1
  else:
    backwardBiCounts[backwardBigram]=1 
  i-=1
for myword in backwardBiCounts:
  splitWords=myword.split(" ")
  firstWord=splitWords[0]
  prob=(backwardBiCounts.get(myword)/uniCounts.get(firstWord))
  backwardBiGramProbs[myword]=prob
print(backwardBiGramProbs)
```
